```python
fname = "../_data/2020-report/scores_with_msas.csv"
```

```python
locations, data = dbna.analysis.parse(fname, selections['sab'])
dist = dbna.analysis.distance_matrix(data)
```

```python
locations, clusters = dbna.analysis.clustering_ap(dist, locations)
dbna.analysis.plot_matrix(locations, dist).set_title('Distance Matrix - Starting a Business Category')
```
