```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sp
from scipy.stats import linregress
```

```python
df = pd.read_csv("../_data/2020-report/scores_with_msas.csv")
```

### Question 1: Correlation between DBNA Score and Relative Business Activity

```python
x_values = df['Activity']
y_values = df['Score']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

slope, intercept, r_value, p_value, std_err = sp.linregress(x_values, y_values)
print(f"The R-Value between DBNA Scores and Relative Business Activity is {r_value}.")

plt.figure(figsize=(14,14))
plt.scatter(x_values,y_values, s=150)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(40,65.5),fontsize=20,color="red")
plt.annotate("Corre. Coeff: ",(40,65.0),fontsize=20,color="red")
plt.annotate(str(round(r_value,5)),(50,65.0),fontsize=20,color="red")
plt.xlabel('Relative Business Activity')
plt.ylabel('Ease of Doing Business Score')
plt.title("DBNA Score vs Business Activity")
plt.show()
```

### Question 2: Clustering Correlation

```python
X = df['Activity'].to_numpy()
_ = df['Score'].to_numpy()
data = np.vstack((X,_)).T
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=15)
kmeans.fit(data)
predicted_clusters = kmeans.predict(data)
```

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.figure(figsize=(14,14))
plt.scatter(X,_, c=predicted_clusters, s=500)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=300, alpha=0.75)
plt.xlabel('Relative Business Activity')
plt.ylabel('Ease of Doing Business Score')
plt.title("DBNA Score vs Business Activity")
plt.show()
```

```python
df = pd.read_csv("../_data/2020-report/cluster_vs_rank.csv")
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111, projection='3d')

x =df['Clusters']
y =df['Activity']
z =df['Score']

ax.scatter(x, y, z, c='r', marker='o', s=200)

ax.set_xlabel('Cluster')
ax.set_ylabel('Relative Business Activity')
ax.set_zlabel('Ease of Doing Business Score')

plt.show()
```

```python
plt.figure(figsize=(14,14))
plt.scatter(df['Clusters'],df['Score'], c=df['Clusters'], s=500)
plt.xlabel('Cluster')
plt.ylabel('Ease of Doing Business Score')
plt.title("Ease of Doing Business Score by Cluster")
plt.show()
```

```python
corr_coef = np.corrcoef(df['Clusters'],df['Score'])[0][1]

plt.figure(figsize=(14,14))
plt.scatter(df['Activity'],df['Score'], c=df['Clusters'], s=600)
plt.annotate("Corre. Coeff: ",(40,65.0),fontsize=20,color="red")
plt.annotate(str(round(corr_coef,4)),(50,65.0),fontsize=20,color="red")
plt.xlabel('Relative Business Activity')
plt.ylabel('Ease of Doing Business Score')
plt.title("DBNA Score vs Business Activity")
print(f"The Correlation Coefficient Between Clusters and EoDB Scores is: {corr_coef}")
plt.show()
```

```python
corr_coef = np.corrcoef(df['Clusters'],df['Score'])[0][1]

plt.figure(figsize=(14,14))
plt.scatter(df['Rank'],df['Score'], c=df['Clusters'], s=600)
plt.xlabel('Ease of Doing Business Rank')
plt.ylabel('Ease of Doing Business Score')
plt.title("Ease of Doing Business Rank & Score")
print(f"The Correlation Coefficient Between Clusters and EoDB Scores is: {corr_coef}")
plt.show()
```
